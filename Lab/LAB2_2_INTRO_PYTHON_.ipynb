```python
#!python
# program fu04.py
# return result from function

#-----------------------------
def Harmonic(n):
    s = 0.0
    i = 1
    while i <= n:
        s = s + 1.0/i
        i = i + 1
    return s

#-----------------------------

print( Harmonic(10) )

print( Harmonic(100) )

print( Harmonic(1000) )

print( Harmonic(1000000) ) # time consuming!!!

print( Harmonic(10000000) ) # time consuming!!!
```

```python
print('For factorial of n (=n!) give the number, n:')
n = int(input())

def factorial(n):
    """Calculate factorial of n (=n!)."""
    a = 1
    while n > 1:
        a *= n
        n -= 1
    return a

print(n, 'factorial =', factorial(n))
```
